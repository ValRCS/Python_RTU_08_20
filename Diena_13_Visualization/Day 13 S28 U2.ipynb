```python
import pandas as pd
```

```python
# # Uzd No.2a
# # https://data.gov.lv/dati/dataset/633b5005-fe7f-44bd-b17c-4995d359f9b0/resource/975fb5a2-1be4-4183-b404-b32afe60c01a/download/kopa_cesu_pils_apmekletaju_veidi_pa_gadiem_menesiem.csv
 

url = 'https://data.gov.lv/dati/dataset/633b5005-fe7f-44bd-b17c-4995d359f9b0/resource/975fb5a2-1be4-4183-b404-b32afe60c01a/download/kopa_cesu_pils_apmekletaju_veidi_pa_gadiem_menesiem.csv'
df = pd.read_csv(url, sep=";") #read csv file from url link with ";" separator
df.head()
```

```python

 
data_year = df[df.Gads == 2015]  # year setup
 
data_year.plot(kind="bar", x="Mēnesis") #standard bar graph with all elements
data_year.plot(x="Mēnesis", y="Ārzemnieku grupa") # line graph with 1 group and all accepteble month if "y" removed than all data will appear
data_year.plot(kind="bar", x="Mēnesis", y=["Ārzemnieku grupa","Latvijas grupa","Ģimene","Pāris"]) # print with History of 4 el
```
