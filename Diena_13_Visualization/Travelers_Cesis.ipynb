```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# tell us what the data is all about
```

```python
df = pd.read_csv('https://data.gov.lv/dati/dataset/633b5005-fe7f-44bd-b17c-4995d359f9b0/resource/052f638a-4394-42b5-b3cc-12fecc3db308/download/kopa_cesu_pils_valstis_pa_gadiem_menesiem.csv', sep = ';')
df
```

```python
filter = (df.gads == 2019) & (df.KOPĀ >= 100) 
filter[:50]
```

```python
df_filtered = df[filter]
df_filtered.sample(5)
```

```python
px.bar(df_filtered, x="Valsts", y="KOPĀ")
```

```python
filter = (df.gads == 2019) & (df.KOPĀ >= 100)  # . syntax works as long as column name has no spaces
# also . syntax columns can not conflict with existing methods
df[filter].plot.bar(y='KOPĀ', x='Valsts' )
```

```python
ukr_visitors = df[df.Valsts == 'Ukraina']

ukr_visitors.plot.line(x='gads', y='KOPĀ', label= 'visitors from Ukraine')
```
