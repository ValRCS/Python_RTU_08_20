```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("../data/kopa_cesu_pils_apmekletaju_veidi_pa_gadiem_menesiem.csv", sep=";")
df
 
#1. Line Plot
 
dfg=df[df['Gads']==2014]
 
dfg.plot(x="Mnesis", y="Skolnu grupa", color="red")
plt.show()
 
#2. Multi-Line Plot
 
dfg=df[df['Gads']==2015]
dfg.plot(x="Mnesis", y=["Skolnu grupa", "imene", "Draugi"] )
plt.show()
 
#3. Stabiu Diagrammu
 
dfg=df[df['Gads']==2016]
dfg.plot.bar(x="Mnesis", y=["Skolnu grupa", "imene", "Draugi"] )
plt.show()
 
#4. Pie-chart
 
labels = 'Janvris', 'Februris', 'Marts', 'Aprlis', 'Maijs'
sizes = [576, 90, 257, 517, 972]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'black']
 
plt.title("Apmekletju skaits 2014 gad")
 
# Plot
plt.pie(sizes, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
 
#5. Stack-Plot - aizpildtu laukuma diagrammu
dfg=df[df['Gads']==2016]
x = dfg["Mnesis"]
 
y1 = dfg["imene"]
y2 = dfg["Draugi"]
 
fig, ax = plt.subplots()
 
ax.stackplot(x, [y1, y2])
 
fig.set_figwidth(12)    
fig.set_figheight(6)    
fig.set_facecolor('floralwhite')
ax.set_facecolor('seashell')
 
plt.show()
```
